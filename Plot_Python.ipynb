```python
###Pie Charts

GDP = (19.4, 11.8 , 4.8, 3.4,2.5,2.4)
countries = ('US' ,'CHINA','JAPAN','GERM','UK','INDIA');


```

```python
# autopct= showing floating data upto how many decimal
# explode= for seperating one of data item like tukada in marathi
plt.figure(figsize=(8,8))
plt.pie(GDP,labels=countries,autopct='%1.02f%%',explode=(0,0.2,0,0,0.1,0.2));
```

```python
# wedgeprops=dict(width=0.5) function is for showing pie chart like donut or Ring
plt.figure(figsize=(8,8))
plt.pie(GDP,labels=countries,autopct='%1.02f%%',wedgeprops=dict(width=0.5));
```
